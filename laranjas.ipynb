```python
import numpy as np

orangeUrl = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'

dados = np.loadtxt(orangeUrl,delimiter=',',skiprows=1,usecols=np.arange(1,6,1))

trans = dados.T

print(trans)
```

```python
grapefruit = trans[0:2,5000:]

grapefruits = grapefruit.T

grapefruits
```

```python
laranja = trans[0:2,:5000]

laranjas = laranja.T

laranjas
```

```python
import matplotlib.pyplot as plt

x = laranjas[:,0]
y = laranjas[:,1]

plt.plot(x,y)
```
